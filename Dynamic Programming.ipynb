# Fibonacci number

```python
def fibb(n,dp):
    if n <= 1:
        return n
    
    if dp[n-1] == -1:
        ans1 = fibb(n-1,dp)
        dp[n-1] = ans1
    else:
        ans1 = dp[n-1]
    if dp[n-2] == -1:
        ans2 = fibb(n-2,dp)
        dp[n-2] = ans2
    else:
        ans2 = dp[n-2]
        
    return ans1 + ans2

# n = int(input())
# dp = [-1 for i in range(n)]
# fibb(n,dp)
```

```python
def fibbI(n):
    dp = [0 for i in range(n+1)]
    dp[0], dp[1] = 0, 1

    for i in range(2,n+1):
        dp[i] = dp[i-1] + dp[i-2]
        
    return dp[n]

# n = int(input())
# fibbI(n)
```

# Min Steps To 1

Given a positive integer 'n', find and return the minimum number of steps that 'n' has to take to get reduced to 1. You can perform any one of the following 3 steps:
1. Subtract 1 from it. (n = n - 1) ,
2. If n is divisible by 2, divide by 2.( if n % 2 == 0, then n = n / 2 ) ,
3. If n is divisible by 3, divide by 3. (if n % 3 == 0, then n = n / 3 ).  

```python
from sys import maxsize as MAX_VALUE

def countMinStepsToOne(n):
    if n == 1:
        return 0
    elif n < 1:
        return

    minSteps = [0]*(n+1)

    for currStep in range(2,n+1):
        substractOne = minSteps[currStep-1]

        divideByTwo = MAX_VALUE
        if currStep%2 == 0:
            divideByTwo = minSteps[currStep//2]

        divideByThree = MAX_VALUE
        if currStep%3 == 0:
            divideByThree = minSteps[currStep//3]

        minSteps[currStep] = 1+min(substractOne,divideByTwo,divideByThree)

    return minSteps[n]

def countMinStepsToOneR(n,dp):
    if n == 1:
        return 0
    elif n < 1:
        return
    
    divideByThree = MAX_VALUE
    if n%3 == 0:
        if dp[n//3] == -1:
            divideByThree = countMinStepsToOneR(n//3,dp)
            dp[n//3] = divideByThree
        else:
            divideByThree = dp[n//3]
            
    divideByTwo = MAX_VALUE
    if n%2 == 0:
        if dp[n//2] == -1:
            divideByTwo = countMinStepsToOneR(n//2,dp)
            dp[n//2] = divideByTwo
        else:
            divideByTwo = dp[n//2]
            
    if dp[n-1]==-1:
        substractOne = countMinStepsToOneR(n-1,dp)
        dp[n-1] = substractOne
    else:
        substractOne = dp[n-1]

    minSteps = 1 + min(substractOne,divideByTwo,divideByThree)
    
    return minSteps
```

# Minimum number of squares

A number can always be represented as a sum of squares of other numbers. Note that 1 is a square and we can always break a number as [(1 * 1) + (1 * 1) + (1 * 1) + â€¦]. Given a number n, find the minimum number of squares that sum to n.

```python
import sys

def minStepsTo1R(n,dp):
    if n == 0:
        return 0

    minSteps = sys.maxsize

    i = 1
    while i*i <= n:
        if dp[n-i*i] != -1:
            currentSteps = 1 + dp[n-i*i]
        else:
            dp[n-i*i] = minStepsTo1R(n-i*i,dp)
            currentSteps = 1+ dp[n-i*i]

        minSteps = min(minSteps,currentSteps)
        i+=1

    return minSteps

def minStepsTo1(n):
    if n == 0:
        return 0

    dp = [0 for i in range(n+1)]

    for i in range(1,n+1):
        dp[i] = sys.maxsize
        for j in range(1,i+1):
            if j*j > i:
                break
            dp[i] = min(dp[i],1+dp[i-j*j])

    return dp[n]
```

# Longest increasing subsequence

Given an array with N elements, you need to find the length of the longest subsequence in the given array such that all elements of the subsequence are sorted in strictly increasing order.

```python
def LIS_R(li,i,n):
    if i == n:
        return 0,0
    
    including_max = 1
    for j in range(i+1,n):
        if li[j] >= li[i]:
            further_including_max = LIS_R(li,j,n)[0]
            including_max = max(including_max,1+further_including_max)
    excluding_max = LIS_R(li,i+1,n)[1]
    
    overallMax = max(including_max,excluding_max)
    
    return including_max,overallMax

def LIS_DP(li,i,n,dp):
    if i == n:
        return 0,0
    
    including_max = 1
    for j in range(i+1,n):
        if li[j] >= li[i]:
            if dp[j] == -1:
                ans = LIS_DP(li,j,n,dp)
                dp[j] = ans
                further_including_max = ans[0]
            else:
                further_including_max = dp[j][0]
            including_max = max(including_max,1+further_including_max)
    if dp[i+1] == -1:
        ans = LIS_DP(li,i+1,n,dp)
        dp[i+1] = ans
        excluding_max = ans[1]
    else:
        excluding_max = dp[i+1][1]
        
    overallMax = max(including_max,excluding_max)
    
    return including_max,overallMax

def LIS_I(li,n):
    
    dp=[[0 for j in range(2) ]for i in range(n+1)]
    
    for i in range(n-1,-1,-1):
        including_max = 1
        for j in range(i+1,n):
            if li[j] >= li[i]:
                including_max = max(including_max,1+dp[j][0])
        dp[i][0] = including_max
        excluding_max = dp[i+1][1]
        
        overallMax = max(including_max,excluding_max)
        dp[i][1] = overallMax
        
    return dp[0][1]
```

# 0 1 Knapsack

A thief robbing a store can carry a maximal weight of W into his knapsack. There are N items, and i-th item weigh 'Wi' and the value being 'Vi.' What would be the maximum value V, that the thief can steal?

```python
from sys import maxsize as MAX_VALUE

# Recursive
def knapsack(W,val,wt,n,i):
    if i == n:
        return 0
    
    if wt[i] > W:
        ans = knapsack(W,val,wt,n,i+1)
    else:
        # Inclusion of ith item
        ans1 = val[i] + knapsack(W-wt[i],val,wt,n,i+1)
        # Exclusion of ith item
        ans2 = knapsack(W,val,wt,n,i+1)
        ans = max(ans1,ans2)
        
    return ans

def knapsack(W,val,wt):
    n = len(val)
    dp = [[0 for j in range(W+1)] for i in range(n+1)]
    
    for i in range(1,n+1):
        for j in range(1,W+1):
            if j < wt[i-1]:
                ans = dp[i-1][j]
            else:
                ans1 = val[i-1] + dp[i-1][j-wt[i-1]]
                ans2 = dp[i-1][j]
                ans = max(ans1,ans2)
            dp[i][j] = ans
            
    return dp[n][W]
```
